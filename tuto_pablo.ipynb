```python
# Load data from CSV
bios = pd.read_csv('./data/bios.csv')
```

```python
filt = bios['height_cm'] >= 220, ['name', 'born_country']

bios.loc[filt]
```

```python
filt = (bios['height_cm'] >= 190) & (bios['born_country'] == 'ARG')

bios.loc[filt]
```

```python
filt = bios['name'].str.contains('Pablo|Antonella', case=False)

bios.loc[filt]
```

```python
bios.query('born_city == "Santa Fe"')
```

```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
bios[['name', 'height_category']].head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 80:
        return 'middleweight'
    
    else:
        return 'heavyweight'
    
bios['Category'] = bios.apply(categorize_athlete, axis=1)
```

```python
bios.head()
```

```python
# Merge DataFrames
nocs = pd.read_csv('./data/noc_regions.csv')
nocs.head()
```

```python
pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left')
```
